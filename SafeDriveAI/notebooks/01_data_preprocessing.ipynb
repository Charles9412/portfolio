```python
from tensorflow.keras.models import load_model

# Load your best checkpoint
model = load_model("../models/saved_model/mobilenetv2_best.h5")

model.summary()
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

img_path = "C:/AI/SafeDriveAI/data/processed/drunk/d4a7a2_1b7fef8b5225436f923db06a76c52e4a_drunk3.jpg"  # Replace with your own
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Normalize

preds = model.predict(img_array)
print(preds)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

val_datagen = ImageDataGenerator(rescale=1.0 / 255)

val_generator = val_datagen.flow_from_directory(
    "../data/processed",         # ✅ same directory used during training
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    shuffle=False                # 🔥 IMPORTANT: do not shuffle, so labels match predictions
)
```

```python
import numpy as np

preds = model.predict(val_generator, verbose=1)
pred_labels = (preds > 0.5).astype(int).flatten()  # binary classification
true_labels = val_generator.classes
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Accuracy
acc = accuracy_score(true_labels, pred_labels)
print(f"Validation Accuracy: {acc:.2%}")

# Detailed report
print(classification_report(true_labels, pred_labels, target_names=["Drunk","Sober"]))

# Confusion matrix
cm = confusion_matrix(true_labels, pred_labels)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Drunk","Sober"], yticklabels=["Drunk","Sober"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import display
import ipywidgets as widgets
import pandas as pd

# Build label_map again
label_map = {v: k for k, v in val_generator.class_indices.items()}

# Create a DataFrame for easy navigation
df_results = pd.DataFrame({
    'Image': val_generator.filepaths,
    'True Label': [label_map[i] for i in true_labels],
    'Predicted Label': [label_map[i] for i in pred_labels]
})
```

```python
def show_image(index, df=df_results):
    row = df.iloc[index]
    img = Image.open(row['Image']).resize((224, 224))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"{row['Image'].split('/')[-1]}\nTrue: {row['True Label']} | Predicted: {row['Predicted Label']}")
    plt.show()
```

```python
from PIL import Image
# Interactive slider for filtered set
image_slider = widgets.IntSlider(min=0, max=len(df_results)-1, step=1, description='Image Index:')
widgets.interact(lambda index: show_image(index, df_results), index=image_slider)
```

```python
import cv2
import mediapipe as mp
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def crop_face(image_path, padding=0.2):
    """
    Detects and crops face using Mediapipe.
    Returns cropped face image (PIL.Image) or None if no face found.
    """
    # Load and convert to RGB
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    h, w, _ = img.shape

    # Initialize Mediapipe face detector
    mp_face_detection = mp.solutions.face_detection
    with mp_face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5) as face_detector:
        results = face_detector.process(img_rgb)

        if results.detections:
            # Get first detected face
            box = results.detections[0].location_data.relative_bounding_box
            xmin = int(box.xmin * w)
            ymin = int(box.ymin * h)
            box_w = int(box.width * w)
            box_h = int(box.height * h)

            # Apply padding
            pad_w = int(box_w * padding)
            pad_h = int(box_h * padding)

            x1 = max(xmin - pad_w, 0)
            y1 = max(ymin - pad_h, 0)
            x2 = min(xmin + box_w + pad_w, w)
            y2 = min(ymin + box_h + pad_h, h)

            face_crop = img_rgb[y1:y2, x1:x2]
            return Image.fromarray(face_crop)

        else:
            print("No face detected.")
            return None
```
